```python
###Project Starts Here

### Time - For Storing Transaction Time
import time


       

###Fetch Inventory From File
fd = open("Inventory.txt","r")
#print(type(fd))
products = fd.read().split('\n')
#print(products)
fd.close()


###Taking User Input
ui_username = input("Enter your Name : ")
ui_phone = input("Enter Phone Number : ")
ui_mail = input("Enter Mail ID : ")
ui_prod_id = input("Enter Product ID :")
ui_prod_qn = int(input("Enter Quantity :"))
print('-'*50)
updated_prod_list = []


#Going through each product detail
for product in products:
    prod_details = product.split(',')
    if ui_prod_id == prod_details[0]: #Check if we have product or not
        if ui_prod_qn <= int(prod_details[3]): #Check If we have enough quantity
            print('-'*20,'TR STORES','-'*19)
            print("Date/Time                  :", time.ctime())
            print("Product Name               :", prod_details[1])
            print("Price                      :", prod_details[2])
            print("Quantity                   :", ui_prod_qn)
            print('-'*50)
            print("Billing Amount             :", ui_prod_qn*int(prod_details[2]))
            print('-'*50)
            prod_details[3] = str(int(prod_details[3]) - ui_prod_qn)

            ###Saving Sales Details
            fd = open('Sales.txt','a')
            Sales_Data = ui_username+','+ui_phone+','+ui_mail+','+ui_prod_id+','+prod_details[1]+','+str(ui_prod_qn)+','+str(ui_prod_qn*int(prod_details[2]))+','+time.ctime()
            fd.write(Sales_Data+'\n')
            fd.close()
        
        else: #If we dont have enough quantity
            print("Sorry, We only have ",prod_details[3])
            print("Would you like to purchase it?")
            print('-'*50)

            ch = input("Press Y/N:")
            if ch == 'Y':
                  print('-'*50)
                  print('-'*20,'TR STORES','-'*19)
                  print("Date/Time                  :", time.ctime())
                  print("Product Name               :", prod_details[1])
                  print("Price                      :", prod_details[2])
                  print("Quantity                   :", prod_details[3])
                  print('-'*50)
                  print("Billing Amount             :", int(prod_details[3])*int(prod_details[2]))
                  print('-'*50)
                  

                  #Saving Sales Details
                  fd = open('Sales.txt','a')
                  Sales_Data = ui_username+','+ui_phone+','+ui_mail+','+ui_prod_id+','+prod_details[1]+','+prod_details[3]+','+str(int(prod_details[3])*int(prod_details[2]))+','+time.ctime()
                  fd.write(Sales_Data+'\n')
                  fd.close()

                  prod_details[3] = '0'
            else:
                print("Thanks")
    updated_prod_list.append(prod_details)  #Add items to Updated List 
        


if updated_prod_list[len(updated_prod_list)-1] == ['']:
    updated_prod_list.pop() #Popping out the [''] created by '\n' below - while writing itms to file

#Displaying Updated List
print('-'*50)
print("Inventory Sucesfully Updated - Check below")
print(updated_prod_list)
print('-'*50)
    

### Writing Updated List Back To File
fd = open("Inventory.txt","w")
for i in updated_prod_list:
        fd.write(i[0] + ','+i[1] +','+i[2]+','+i[3]+'\n') ###This last new line should be removed as its not required after the last itm
fd.close()

#Further Scope - Can also add valiation checks for email, phone, quantity>0,etc
```
